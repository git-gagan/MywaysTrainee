```python
import pandas as pd

banglore_data = pd.read_csv("E:\MywaysProjects\Minor\Requirements and Data Minor Projects (1)\RequirementsNData\data_files\Bengaluru_House_Data.csv")

banglore_data
```

```python
#let's first remove NaN values

banglore_data.dropna(inplace = True)
banglore_data.index+=1
banglore_data   #After removing all the rows with undefined values
```

```python
#let's analyse the property which is ready to move now
banglore_ready = banglore_data[banglore_data["availability"] == "Ready To Move"]
banglore_ready
```

```python
#Plotting area type vs price to find the most costly ready to move properties
df = banglore_ready.groupby("area_type", as_index = False)
df.first()
```

```python
# let's calculate total number of properties in each group
import numpy as np

g1 = df["area_type"].aggregate(len)
g1.index = ["Built up area","Carpet area","Plot area","Super built up area"]
g1
# showing all types available
```

```python
import matplotlib.pyplot as plt

plt.plot(g1.index,g1["area_type"], color = "red")
plt.scatter(g1.index,g1["area_type"])
plt.title("Number of each property type available")
plt.xlabel("Property type")
plt.ylabel("Number")
plt.show()
```

```python
#LET'S calculate the average price of each area group wise
g2 = df["price"].aggregate(np.sum)
g2["area_type_number"] = [i for i in g1["area_type"]]

g2["price_per_unit"] = g2["price"]/g2["area_type_number"]
g2
```

```python
plt.scatter(g2["area_type"],g2["price_per_unit"], color ="red")
plt.plot(g2["area_type"],g2["price_per_unit"])
plt.xlabel("plot type")
plt.ylabel("price_per_unit")
plt.show()
```

```python
# Let's find the biggest ready to move houses

#sq ft data had string values which needs to be handled and specially dealt with..
# done as follows
sqft = []
cols = []
price = []
for i in banglore_ready.index:
    try:
        sqft.append(float(banglore_ready["total_sqft"][i]))
        cols.append(banglore_ready["area_type"][i])  
        price.append(banglore_ready["price"][i])
    except:
        continue       

new_df = pd.DataFrame(sqft, columns = ["sqft"])
new_df["Names"] = cols
new_df["price"] = price

#checking for houses with size greater than 2000

new_df = new_df[new_df["sqft"] > 2000]
new_df
```

```python
plt.hist(new_df["Names"], bins = 4, color = "black")
plt.grid()
```

```python
new_df_grouped = new_df.groupby("Names")

d ={}
for i,v in new_df_grouped.groups.items():
    d[i] = max(v)

biggest_properties = pd.DataFrame(d, index = ["Area"])

biggest_properties

### These are the biggest properties
```

```python
# Let's find if there's a linear relationship between between bath vs price

import seaborn as sns
plt.figure(figsize = (10,6))
plt.xlim(0,2000)
plt.ylim(0,15000)
sns.regplot(x = "price", y = "sqft", data = new_df, color = "purple")
```

```python
#checking correlation

from scipy.stats import pearsonr
pearsonr(new_df["price"],new_df["sqft"])
```
