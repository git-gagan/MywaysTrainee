## Indian Tourist Data Analysis


```python
# Data from https://www.kaggle.com/debayankar7/stateutwise-domesticforeign-tourist-india

import pandas as pd

tourist_data = pd.read_csv("E:\FreeCodeCamp\PythonProjects\Minor Projects\Tourist Data\Data.csv", index_col = 0)

total = tourist_data[-1:]
tourist_data.drop(["Total"], axis= 0, inplace = True)
tourist_data
```

```python
tourist_data.info()

#Tells about the basic info of the dataset
```

```python
# to get number of rows and columns, use shape

tourist_data.shape

#suggesting there are 36 rows and 6 columns
```

```python
# to check for null values

tourist_data.isnull().sum()

# We can conclude that there are no NULL values, that is Data set is clean
```

```python
# Let's combine the domestic values of 2016 and 2017 for each state/UT and get mean domestic
combined_domestic = pd.DataFrame(tourist_data["2016 - Domestic"]  + tourist_data["2017 - Domestic"])

combined_domestic.columns = ["16-17combined"]

combined_domestic
```

```python
# Similarily let's just plot it for foreign combined
combined_foreign = pd.DataFrame(tourist_data["2016 - Foreign"]  + tourist_data["2017 - Foreign"])

combined_foreign.columns = ["16-17combined"]

combined_foreign
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.figure(figsize = (12,6))
plt.xlabel("States")
plt.ylabel("Tourists")
plt.title("Combined Tourism Domestic + Foreign")
plt.xticks(rotation = "vertical")
plt.yscale("log")
plt.scatter(combined_domestic.index, combined_domestic["16-17combined"], s = tourist_data["Growth Rate - Domestic"].abs()*4)
plt.scatter(combined_foreign.index,combined_foreign["16-17combined"], s = tourist_data["Growth Rate - Foreign"].abs()*4, c ="Red")
plt.legend(["Domestic","Foreign"], loc= "lower right")
plt.show()
```

```python
# get the states/UT with positive negative rate in both tourisms

for index in tourist_data[np.logical_and(tourist_data["Growth Rate - Domestic"] < 0, tourist_data["Growth Rate - Foreign"] < 0)].index:
    print(index)
```

```python
#Prepare a final visualisation of 2016 vs 2017
tourist_data.rename(columns ={"2016 - Foreign": "16f","2017 - Foreign": "17f","2016 - Domestic": "16d","2017 - Domestic": "17d"}, inplace = True)

tourist_data.head(3)   #renamed
```

```python
combined_16 = tourist_data["16f"] + tourist_data["16d"]
combined_16 = pd.DataFrame(combined_16)

combined_16.columns = ["2016"]

combined_17 = tourist_data["17f"] + tourist_data["17d"]
combined_17 = pd.DataFrame(combined_17)

combined_17.columns = ["2017"]

print(combined_16)
print(combined_17)

print(combined_16["2016"] < combined_17["2017"])
```

```python
#finally plot a visualisation
plt.figure(figsize = (12,8))
plt.scatter(combined_16.index,combined_16["2016"], c = "red")
plt.scatter(combined_17.index,combined_17["2017"], c = "black")
plt.legend(["2016","2017"], loc = "lower right")
plt.xlabel("States/UT")
plt.yscale("log")
plt.xticks(rotation = 90)
plt.ylabel("Tourists")
plt.title("2016 vs 2017")
plt.grid(True)
plt.show()
```

```python
# Final visualisation of growth rate domestic vs foreign

plt.figure(figsize = (14,7))
# used line formatting as 3rd argument below (o-g) --> marker(o), line(-), color(green)
plt.plot(tourist_data[["Growth Rate - Domestic"]].index, tourist_data["Growth Rate - Domestic"],"o-b")
plt.plot(tourist_data[["Growth Rate - Foreign"]].index, tourist_data["Growth Rate - Foreign"],"*-g")
plt.xticks(rotation = 90)
plt.legend(["Domestic Rate","Foreign Rate"])
plt.xlabel("States")
plt.ylabel("Growth %")
plt.grid(True)
plt.title("Domestic vs Foreign Growth Rate")
plt.show()
```

```python
# States with foreign growth rate greater than domestic growth rate
tourist_data[tourist_data["Growth Rate - Domestic"] - tourist_data["Growth Rate - Foreign"] < 0]
```

```python
print("State with max difference :",(tourist_data["Growth Rate - Foreign"] - tourist_data["Growth Rate - Domestic"]).abs().idxmax())
#idxmax() --> for id of max value
print("State with min difference :",(tourist_data["Growth Rate - Foreign"] - tourist_data["Growth Rate - Domestic"]).abs().idxmin())
```
